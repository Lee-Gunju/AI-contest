```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

path = "data.csv"
data = np.genfromtxt(path, delimiter=',')

x_data = data[:, 0]
y_data = data[:, 1]

plt.figure(figsize=(8, 8))
plt.scatter(x_data, y_data, alpha=0.3)
plt.show()
```

```
w = -30
b = -30

n_data = len(x_data)
epochs = 600
learning_rate = 0.01 # step_size 
w_list = []
b_list = []
cost_list = []

for i in range(epochs):
    hypothesis = x_data * w + b
    cost = np.sum((hypothesis - y_data) ** 2) / 2 * n_data  # objective function 
    gradient_w = np.sum((w * x_data - y_data + b) * 2 * x_data) / n_data # gradient descent θ1
    gradient_b = np.sum((w * x_data - y_data + b) * 2) / n_data # gradient descent θ0 

    w -= learning_rate * gradient_w # gradient descent θ1
    b -= learning_rate * gradient_b # gradient descent θ0

    w_list.append(w)
    b_list.append(b)
    cost_list.append(cost)

plt.plot(x_data, y_data, 'ko', label = 'data') # 2.Output results
plt.plot(x_data, hypothesis, 'r', label = 'solution graph') # 2.Output results
plt.legend()
plt.show()
```
